### Parking Analysis

```python
midpoint_gps = load_midpoint_id()
midpoint_gps = midpoint_gps[['lat_mid','lng_mid']].as_matrix()
train_streets = [col.replace(' ','').lower() for col in train_df['Street'].unique()]
```

```python
parking_df = pd.read_feather('../ref_data/gmaps_df_parking.feather')
parking_df['STREET'] = parking_df['street_name'].map(lambda x : x.replace(' ','').lower())
parking_df = parking_df[parking_df['STREET'].isin(train_streets)]
parking_df['date'] = pd.to_datetime(parking_df['datetime'], unit='ms', utc=True)

parking_df['hour'] = parking_df['date'].dt.hour
parking_df['dow'] = parking_df['date'].dt.dayofweek
parking_df['month'] = parking_df['date'].dt.month
parking_df['day'] = parking_df['date'].dt.day
parking_df['year'] = parking_df['date'].dt.year
```

### Calculating Haversine distance

```python
parking_pid = pd.DataFrame(parking_df.groupby(['pid','lat','lon']).size()).reset_index()
parking_pid = parking_pid[['lat','lon']].as_matrix()
```

```python
distances = np.zeros((len(parking_pid), len(midpoint_gps)))
for i, row in enumerate(parking_pid):
    for j, col in enumerate(midpoint_gps):
        distances[i,j] = haversine(row,col, miles=True)
```

```python
distances_selection = np.argmin(distances,axis=1) 
distances_min = np.min(distances,axis=1) 
```

```python
pid2loc_id = pd.DataFrame({'loc_id':distances_selection , 'lowest_dist_ft':distances_min*5280})
pid2loc_id['pid'] = pid2loc_id.index
pid2loc_id = pid2loc_id[pid2loc_id['lowest_dist_ft'] <300]
pid2loc_id.hist('lowest_dist_ft')
```
